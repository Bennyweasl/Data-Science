### Latent Semantic Analysis
<l>LSA on a group of newsgroup posts from the newsgroup 'rec.sport.baseball.'

```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_20newsgroups
categories = ['rec.sport.baseball']
dataset = fetch_20newsgroups(subset='all',shuffle=True, random_state=42, categories=categories)
corpus = dataset.data
```

### Code from https://github.com/mbernico/CS570.git

```python
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
#run this only once
nltk.download('stopwords')
```

```python
stopset = set(stopwords.words('english'))
```

```python
vectorizer = TfidfVectorizer(stop_words=stopset,
                                 use_idf=True, ngram_range=(1, 3))
```

```python
X = vectorizer.fit_transform(corpus)
```

```python
X[0]
```

```python
print X[0]
```

```python
X.shape
```

<li> That is a massive dataset

```python
lsa = TruncatedSVD(n_components=30, n_iter=100)
lsa.fit(X)
```

<li>n_components : int, default = 2
Desired dimensionality of output data. Must be strictly less than the number of features. The default value is useful for visualisation. For LSA, a value of 100 is recommended.
<li>n_iter : int, optional
Number of iterations for randomized SVD solver. Not used by ARPACK.

```python
lsa.components_[0]
```

```python
import sys
print (sys.version)
```

<Li> displaying results of TF-IDF in concepts groups with listed out to 10

```python
terms = vectorizer.get_feature_names()
for i, comp in enumerate(lsa.components_): 
    termsInComp = zip (terms,comp)
    sortedTerms =  sorted(termsInComp, key=lambda x: x[1], reverse=True) [:10]
    print "Concept %d:" % i
    for term in sortedTerms:
        print term[0]
    print " "
```
